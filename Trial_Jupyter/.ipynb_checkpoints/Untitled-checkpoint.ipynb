```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import animation


# plotting 3D complex plane
from mpl_toolkits.mplot3d import Axes3D
import mpl_toolkits.mplot3d.axes3d as p3

import site
site.addsitedir('..\LabCommonLib')
from   pyLabDSP import myFnSig
```

```python
# Lets create the data
A=0.98; w1=0.1*np.pi; Phi = np.pi/6;  numSamples = 45
[n,y1] = myFnSig.fnGenDTComplexExp(A, w1, Phi, numSamples)

```

```python
#plotting in 2-d, both the real and imag values
# first figure is real (red) and imaginary (green) lines vs time (samples)
plt.figure()
plt.plot(n, y1[0:numSamples].real,'r--o')
plt.plot(n, y1[0:numSamples].imag,'g--o')
plt.xlabel('sample n'); plt.ylabel('y[n] (real/imag)')
plt.grid()
plt.title('(red=real  and green=imag)  vs time')
plt.show()
```

```python
# Lets plot in polar
#plotting in polar
plt.figure()
for x in y1:
    plt.polar([0,np.angle(x)],[0,np.abs(x)],marker='o')
plt.show()

```

```python
# Lets animiate in polar
plt.figure()
ax = plt.polar()
fig = plt.figure()

# animation function.  This is called sequentially
def animate(i):
    plt.polar([0,np.angle(y1[i])],[0,np.abs(y1[i])],marker='o')

# call the animator.  blit=True means only re-draw the parts that have changed.
ani = animation.FuncAnimation(fig, animate, frames=len(n))
ani
```

```python
# plotting 3D complex plane
plt.figure()
fig = plt.figure()
ax = fig.gca(projection='3d')
plt.rcParams['legend.fontsize'] = 10

reVal = y1[0:numSamples].real
imgVal = y1[0:numSamples].imag
ax.plot(n,reVal, imgVal,  label='complex exponential phasor')
ax.scatter(n,reVal,imgVal, c='r', marker='o')
ax.set_xlabel('sample n')
ax.set_ylabel('real')
ax.set_zlabel('imag')
ax.legend()
plt.show()

```

```python
# plotting 3D complex plane

# Attaching 3D axis to the figure
plt.figure()
fig = plt.figure()
ax  = p3.Axes3D(fig)
ax.set_xlim3d(0, len(n))
ax.set_ylim3d(-1.2,1.2)
ax.set_zlim3d(-1.2,1.2)

plt.rcParams["animation.html"] = "jshtml"
plt.rcParams['legend.fontsize'] = 10

def animate(i):
    ax.plot(n[:i],y1[:i].real, y1[:i].imag,label='complex exponential')
    ax.scatter(n[:i],y1[:i].real, y1[:i].imag, c='r', marker='o')
    
    
ani = animation.FuncAnimation(fig, animate, frames=len(n))
ani


```
