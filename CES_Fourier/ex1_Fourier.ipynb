```python
# This is to demonstrate python 
# generating signals, one a (pretend) CT version, that is sampled 100 times more
# than its Discrete time counter part

# Name: Chng Eng Siong
# Date: Oct 2018

import numpy as np
import matplotlib.pyplot as plt


A = 1
Fs    = 10
Freq  = 1
Phi = 0
sTime = 0
eTime = 2
tCT = np.arange(sTime,eTime,1.0/(Fs*100))
yCT = A*np.cos(2 * np.pi * Freq * tCT + Phi)

tDT = np.arange(sTime,eTime,1.0/Fs)
yDT = A*np.cos(2 * np.pi * Freq * tDT + Phi)

# Lets plot what we have
plt.figure(1)
plt.plot(tCT, yCT,'r-');
plt.plot(tDT, yDT,'go');
plt.xlabel('t'); plt.ylabel('y(t)')
plt.grid()
plt.show()

```

# Explanantion of the above plot
The above is a plot comparing a CT (continuous time) signal vs
a dicrete time signal. The red line denotes the continuous time.
the green dots the sampled sequence.


```python
# Lets now generate a foueir basis
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D


A = 1
Fs    = 10
Freq  = 1
Phi = 0
sTime = 0
eTime = 2
tCT = np.arange(sTime,eTime,1.0/(Fs*100))
yCT = A*np.cos(2 * np.pi * Freq * tCT + Phi)

tDT = np.arange(sTime,eTime,1.0/Fs)
yDT = A*np.cos(2 * np.pi * Freq * tDT + Phi)


N = len(tCT)
k = 1
y1 = A*np.exp(-1j*(2 * np.pi/N*k * tCT))
numSamples = len(y1)

# Lets plot what we have
# plotting 3D complex plane
plt.rcParams['legend.fontsize'] = 10
fig = plt.figure()
ax = fig.gca(projection='3d')
reVal = y1[0:numSamples].real
imgVal = y1[0:numSamples].imag
ax.plot(tCT,reVal, imgVal,  label='complex exponential phasor')
ax.scatter(tCT,reVal,imgVal, c='r', marker='o')
ax.set_xlabel('sample n')
ax.set_ylabel('real')
ax.set_zlabel('imag')
ax.legend()
plt.show()

                     
```
